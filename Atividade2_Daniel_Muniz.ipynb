```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_excel('/content/default_credit1.xls')
df.head()

#Questão 1
# Características de valor da fatura
valor_fatura = ['BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3',
                 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6']

# Características de pagamento
faturas_caract = ['PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3',
                'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6']

print("Características de valor da fatura:", valor_fatura)
print("Características de pagamento:", faturas_caract)

#Questão 2 - Não fez muito sentido, entendi um pouco das linhas apenas
df[valor_fatura].describe()

#Questão 3
plt.figure(figsize=(12, 8))
for i, col in enumerate(valor_fatura):
    plt.subplot(2, 3, i + 1)
    plt.hist(df[col], bins=20)
    plt.title(col)
plt.tight_layout()
plt.show()

#Questão 4 - Uma linha com valores em 0 e o count mesmo valores nas 2 leituras
df[faturas_caract].describe()

#Questão 5 -
plt.figure(figsize=(12, 8))
for i, col in enumerate(faturas_caract):
    plt.subplot(2, 3, i + 1)
    plt.hist(df[col], bins=20)
    plt.title(col)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

#Questão 6
(df[faturas_caract] == 0).sum()

#Questão 7

# Máscara para valores diferentes de zero
mask = df[faturas_caract] != 0

# log10 apenas nos valores não nulos
df_log = df[faturas_caract][mask].apply(np.log10)

# Histogramas logarítmicos
plt.figure(figsize=(12, 8))
for i, col in enumerate(faturas_caract):
    plt.subplot(2, 3, i + 1)
    plt.hist(df_log[col].dropna(), bins=20)
    plt.title(f'log10({col})')
plt.tight_layout()
plt.show()


```
